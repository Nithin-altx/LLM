```python
import os
from dotenv import load_dotenv
load_dotenv()
ollama_key=os.getenv("LANGCHAIN_API_KEY")
```

```python
from langchain_ollama import ChatOllama
llm= ChatOllama(
     model="llama2"
)
response=llm.invoke("Suggest a movie title to watch")
response.content
```

```python
messages = [
    (
        "system",
        "You are a helpful assistant that translates English to hindi. Translate the user sentence.",
    ),
    ("human", "I love programming."),
]
ai_msg = llm.invoke(messages)
ai_msg
```

```python
print(ai_msg.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate


prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful programming assistant."),
    ("human", "Explain {concept} with an example in {language}.")
])

messages = prompt.format_messages(
    concept="prompt templates",
    language="Python"
)

print(messages)
```

```python
response=llm.invoke(messages)
```

```python
response.content
```

```python
#Using prompt template
from langchain_core.prompts import PromptTemplate
```

```python
pt=PromptTemplate.from_template("who is the father of {Author}")
prompt=pt.invoke({"Author":"AI"})
res=llm.invoke(prompt)
res
```

```python
#Chaining refers to sequence of tasks grouping together called runnable
from langchain_core.output_parsers import StrOutputParser
prompt2=pt | llm | StrOutputParser()
res=prompt2.invoke({"Author":"Biology"})
```

```python
print(res)
```
